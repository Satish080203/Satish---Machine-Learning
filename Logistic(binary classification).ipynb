```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_excel("LR.xlsx")
df.head()
```

```python
plt.scatter(df.age,df.insurance,color='red',marker='+')
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(df[['age']],df.insurance,test_size=0.2,random_state=10)
```
