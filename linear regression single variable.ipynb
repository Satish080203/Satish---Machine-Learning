```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
df=pd.read_csv("linear.csv")
df
```

```python
df.size
```

```python
df.describe()
```

```python
df.shape
```

```python
df.head()
```

```python
plt.scatter(df.area,df.price,color="red",marker="*")
plt.xlabel("AREA")
plt.ylabel("PRICE")
plt.title("AREA VS PRICE")
plt.show()
```

```python
reg=LinearRegression()
```

```python
reg
```

```python
reg.fit(df[['area']],df.price)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
reg.predict([[1000]])
```

```python
reg.predict([[5800]])
```

```python
plt.scatter(df.area,df.price,color="red",marker="*")
plt.plot(df.area,reg.predict(df[['area']]),color="blue")
plt.xlabel("AREA")
plt.ylabel("PRICE")
plt.title("AREA VS PRICE")
plt.show()
```

```python
import pickle
```

```python
with open('model_pickle','wb')as f:
    pickle.dump(reg,f)
```

```python
with open('model_pickle','rb')as f:
    mp=pickle.load(f)
```

```python
mp.predict([[5000]])
```
